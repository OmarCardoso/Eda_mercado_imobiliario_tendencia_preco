```python
mel_data = mel_data.dropna(axis=0)
```

```python
# cria variÃ¡vel com serie de preÃ§os
y = mel_data.Price
```

```python
y
```

```python
# seleciona colunas
mel_data_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
```

```python
# coloca as colunas do dataset em x
X = mel_data[mel_data_features]
```

```python
# Importa a biblioteca DecisionTreeeRegressor
from sklearn.tree import DecisionTreeRegressor

# Define modelo. Especifica o numero de random_state para termos o mesmo resultado qd executado
melbourne_model = DecisionTreeRegressor(random_state=1)

# Ajusta modelo
melbourne_model.fit(X, y)
```

```python
# previsÃµes da primeiras linhas para verificar como as funcoes funcionam
print ("Fazendo previsÃµes para as 5 primeiras casas:")
print(X.head())
print("As previsÃµes sÃ£o")
print(melbourne_model.predict(X.head()))
```
